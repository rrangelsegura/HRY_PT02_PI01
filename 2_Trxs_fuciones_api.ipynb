```python
import pandas as pd

movies = pd.read_csv('Datos/movies_depurado.csv')

#Generos
genres_x_movie = pd.read_csv('Datos/generos_x_movie.csv')
genres_master = pd.read_csv('Datos/generos_x_movie.csv')


#productoras
prods_x_movie = pd.read_csv('Datos/productoras_x_movie.csv')
prods_master = pd.read_csv('Datos/mestro_productoras.csv')

#paises
coutries_x_movie = pd.read_csv('Datos/paises_x_movie.csv')
pcts_master = pd.read_csv('Datos/maestro_paises.csv')


#idiomas
langs_x_movie = pd.read_csv('Datos/idiomas_x_movie.csv')
langs_master = pd.read_csv('Datos/maestro_idiomas.csv')


```

```python
#Transformacion peliculas_idioma
nro_peliculas_x_idioma = movies[['movie_id','original_language']].groupby('original_language').count()
nro_peliculas_x_idioma.rename(columns={"movie_id": "cantidad"}, inplace=True)
nro_peliculas_x_idioma.sort_values(by=['cantidad'], ascending=False, inplace=True)
nro_peliculas_x_idioma.to_csv('Datos/nro_peliculas_x_idioma.csv',',',index=False)
```

```python
#Funcion peliculas_idioma
def peliculas_idioma(idioma:str):
    '''Recibe la abreviatura del idioma y retorna la cantidad de peliculas producidas en ese idioma'''
    return nro_peliculas_x_idioma[nro_peliculas_x_idioma.index == idioma]

#prueba
peliculas_idioma('fr')
```

```python
# Transfomacion peliculas_duracion
drcion_anio_x_peli = movies[['title','runtime','release_year']]
drcion_anio_x_peli.to_csv('Datos/drcion_anio_x_peli.csv',',',index=False)
```

```python
# Funcion peliculas_duracion
def peliculas_duracion(pelicula:str):
    '''Recibe el nombre de una pelicula y devuelve la duracion y el año'''
    return drcion_anio_x_peli[drcion_anio_x_peli['title'] == pelicula]

#prueba
peliculas_duracion('Wife')
```

```python
# Transformacion franquicia
nro_pels_x_franq = movies[['movie_id','belongs_to_collection']].groupby('belongs_to_collection').count()
nro_pels_x_franq.rename(columns={"movie_id": "nro_peliculas"}, inplace=True)

totrev_x_franq = movies[['revenue','belongs_to_collection']].groupby('belongs_to_collection').sum()
totrev_x_franq.rename(columns={"revenue": "total_revenue"}, inplace=True)
avgrev_x_franq = movies[['revenue','belongs_to_collection']].groupby('belongs_to_collection').mean()
avgrev_x_franq.rename(columns={"revenue": "average_revenue"}, inplace=True)

nro_pels_totrev_x_franq = nro_pels_x_franq.merge(totrev_x_franq,how = 'left', on = 'belongs_to_collection')
nro_pels_avgrev_totrev_x_franq = nro_pels_totrev_x_franq.merge(avgrev_x_franq,how = 'left', on = 'belongs_to_collection')
nro_pels_avgrev_totrev_x_franq.sort_values(by=['nro_peliculas'], ascending=False, inplace=True)
nro_pels_avgrev_totrev_x_franq.to_csv('Datos/nro_pels_avgrev_totrev_x_franq.csv',',',index=False)
```

```python
# Funcion franquicia
def franquicia(franquicia:str):
    '''Recibe el nombre de la franquicia y retorna el numero total de peliculas, la ganacia total y la ganancia promedio'''
    return nro_pels_avgrev_totrev_x_franq[nro_pels_avgrev_totrev_x_franq.index==franquicia]

#prueba
franquicia('James Bond Collection')
```

```python
# Transformacion peliculas_pais
movies_x_country = coutries_x_movie[['pais','movie_id']].groupby('pais').count()
movies_x_country.rename(columns={"movie_id": "cantidad"}, inplace=True)
movies_x_country.to_csv('Datos/movies_x_country.csv',',',index=False)
```

```python
# Funcion peliculas_pais
def peliculas_pais(pais:str):
    '''Recibe el nombre del pais y retorna la cantidad de peliculas producidas en dicho pais'''
    return movies_x_country[movies_x_country.index==pais]

# prueba
peliculas_pais('Colombia')    
```

```python
# Transformacion productoras_exitosas
revenue_x_productora = prods_x_movie[['productora','revenue']].groupby('productora').sum()
pelis_x_productora = prods_x_movie[['productora','movie_id']].groupby('productora').count()
pelis_x_productora.rename(columns={"movie_id": "nro_peliculas"}, inplace=True)
pelis_rev_x_productora = revenue_x_productora.merge(pelis_x_productora,on = 'productora', how = 'left')
pelis_rev_x_productora.to_csv('Datos/pelis_rev_x_productora.csv',',',index=False)
```

```python
pelis_rev_x_productora.sort_values(by=['nro_peliculas'], ascending=False, inplace=True)
pelis_rev_x_productora
```

```python
# Funcion productoras_exitosas
def productoras_exitosas(productora:str):
    '''Recibe el nombre de una productora y retorna el revenue total de la productora y la cantidad de peliculas que realizo'''
    return pelis_rev_x_productora[pelis_rev_x_productora.index == productora]

# Prueba
productoras_exitosas('Warner Bros.')
```
